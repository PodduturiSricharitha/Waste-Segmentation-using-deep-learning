```python
from roboflow import Roboflow
rf = Roboflow(api_key="ciacgAx81gkETPCt8Ukw")
project = rf.workspace("charitha-ppjnc").project("waste-detection-2-whu8d")
version = project.version(1)
dataset = version.download("yolov8")
```

```python
from ultralytics import YOLO
# Load the pre-trained YOLOv8 model (transfer learning from COCO)
model = YOLO("yolov8l-seg.pt")

# Train the model 
model.train(
    data=f"{dataset.location}/data.yaml",  # Point to your data.yaml file
    epochs=100,                            # Train for 100 epochs
    imgsz=640,                             # Image size
    batch=16,                              # Batch size
    name="custom_yolo_model",              # Model name for saving
    patience=20,                           # Early stopping patience
    device="0",                            # Use first GPU
    augment=False                          
)

# Evaluate the model
metrics = model.val()
print("Validation Metrics:", metrics)

# Export the model to save it
model.export(format='onnx')  
```
